```python
#!/usr/bin/env python3
import os
import ethercalc
import datetime
import pytz

sheet_id = 'ru0aesdkj7pa'
e = ethercalc.EtherCalc("https://ethercalc.net")

for i in range(1,100):
    status = e.cells(sheet_id, "S"+str(i))["datavalue"]
    print(status)
    if status == 'w':
        d = e.cells(sheet_id, "A"+str(i))["datavalue"]
        D = e.cells(sheet_id, "B"+str(i))["datavalue"]
        r = e.cells(sheet_id, "C"+str(i))["datavalue"]
        m_1 = e.cells(sheet_id, "D"+str(i))["datavalue"]
        h = e.cells(sheet_id, "E"+str(i))["datavalue"]
        X = e.cells(sheet_id, "F"+str(i))["datavalue"]
        PARA = e.cells(sheet_id, "G"+str(i))["datavalue"]
        
        str_d = str(d)
        str_D = str(D)
        str_r = str(r)
        str_m_1 = str(m_1)
        str_h = str(h)
        str_X = str(X)
        str_PARA = str(PARA)
        
        output_file = "../output/out_"+str_d+"_"+str_X+"_"+str_PARA+".ipynb"
        parameters = " -p d "+ str_d+"  -p D "+ str_D+" -p r " + str_r+" -p m_1 " + str_m_1+" -p h " + str_h+" -p X " + str_X+" -p PARA " + str_PARA
        cmd = ["papermill", "real_quadratic_narrow_principal_papermill_template.ipynb", output_file, parameters,  "-k sagemath", "--autosave-cell-every 30" , "--progress-bar"]
        str_cmd = '  '.join(cmd)
        print(str_cmd)
        e.command(sheet_id, [ethercalc.set("S"+str(i), "'r")])
        os.system(str_cmd)
        e.command(sheet_id, [ ethercalc.set("S"+str(i), "'d")])
        e.command(sheet_id, [ ethercalc.set("T"+str(i), "'" + str(datetime.datetime.now(tz=pytz.timezone("Asia/Tokyo"))))])
        
```
